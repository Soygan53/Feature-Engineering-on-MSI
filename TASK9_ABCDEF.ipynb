 TASK-9A: Please apply the following feature engineering steps on the numeric features of the AD dataset.

 Divide by NACCBRNV (MRI)
 Multiply by NACCAGEB


```python
import pandas as pd
import numpy as np

data = pd.read_excel("Temporary_data3_Left_Right_Copy.xlsx")
data.head()
```

```python
maske = ['float64']
data_new = data.select_dtypes(include=maske)
data_new
```

```python
Naccbrnv = data["NACCBRNV"]
Naccageb = data["NACCAGEB"]

data_new = data_new.div(Naccbrnv, axis=0) #divide by
data_new = data_new.mul(Naccageb, axis=0) #multiple by
data_new
```

###  TASK-9B: Please apply Gaussian Naive Bayes algorithm on the numeric values (after the transformation) to predict the CDRGLOB values (class-based). Train-test split, train by grid search (k=5) and optimize for the parameter values.

```python
from sklearn.model_selection import train_test_split

Cdrglob= data["CDRGLOB"]
X_train, X_test, y_train, y_test = train_test_split(data_new, Cdrglob, test_size=0.2, random_state=42)
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(X_train, y_train)
y_predicted = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_predicted)
```

```python
from sklearn.model_selection import GridSearchCV
param_grid = {'var_smoothing': np.logspace(0,-9, num=100)}
grid = GridSearchCV(model, param_grid, cv=5, scoring='accuracy')
grid.fit(X_train, y_train)

print("Best parameter: ", grid.best_params_)
print("Score: ",grid.score(X_test, y_test))
```

###  TASK-9E: Please merge the normalized numeric values and one-hot encoded categorical values and then apply the following regression methods (Linear, Ridg and Lasso) to predict the labels as numeric values. Please do not forget to optimize them and use R2 for scoring.

```python
data_merged=pd.merge(data,encoded_data)
data_merged.head()
```

```python
from sklearn.metrics import r2_score
from sklearn.linear_model import LinearRegression, Ridge, Lasso

X = data_merged.drop(columns="CDRGLOB")

Xtrain, Xtest, Ytrain, Ytest = train_test_split(X, Cdrglob, test_size=0.2, random_state=0)

linear_model = LinearRegression()
linear_predicted = linear_model.fit(Xtrain,Ytrain).predict(Xtest)

print("Linear score :",r2_score(Ytest, linear_predicted))

ridge_model = Ridge(alpha=0.1)
ridge_predicted = ridge_model.fit(Xtrain,Ytrain).predict(Xtest)

print("Ridge score : ",r2_score(Ytest, ridge_predicted))

lasso_model = Lasso(alpha=0.1)
lasso_predicted = lasso_model.fit(Xtrain,Ytrain).predict(Xtest)

print("Lasso score : ",r2_score(Ytest, lasso_predicted))
```
